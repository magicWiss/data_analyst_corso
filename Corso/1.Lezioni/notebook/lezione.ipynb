```python
#prodotto matriciale
A=np.array([
        [1,2,3],
        [4,5,6],
        [7,8,10]])
B=[1,2,3]

#metodo 1 (build in funciton)
C=np.matmul(A,B)
C
```

```python
#metodo 2 (utilizzo di @)
D=A @ B
D
```
